## 2. The question: decision, action, cost of a wrong call

**Research Question:**
Can machine learning identify web pages that are likely to decline in search performance so that content teams can prioritize which pages to refresh first?

**Decision:**
Which pages should be refreshed first?

**Action:**
Refresh the highest-priority pages before they lose more search traffic.

**Cost of a wrong recommendation:**
If a declining page is missed, search traffic may continue to decrease. If a healthy page is refreshed unnecessarily, time and resources are wasted.

**Why ML helps:**
Machine learning can combine multiple signals such as impressions, CTR, content age, and average position to identify patterns that are difficult to capture using simple manual rules.

## 3. Quick look at the data (2–3 real numbers)

The starter dataset contains **30,000 pages**.

The trend distribution is:
- Down: **16,262**
- Stable: **5,962**
- Up: **4,388**
- New: **2,236**
- Flat: **1,152**

The median word count is similar across the trend categories (around **2,700–2,900 words**), suggesting that content length alone is not enough to explain whether a page grows or declines.

These observations indicate that multiple factors influence page performance, making this a suitable machine learning problem.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))

print("\nTrend direction counts:")
print(df["trend_direction"].value_counts())

print("\nMedian word count by trend:")
print(df.groupby("trend_direction")["word_count"].median().round(0))
```
